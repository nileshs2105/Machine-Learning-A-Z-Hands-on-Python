# SIMPLE LINEAR REGRESSION

### Formula: y = b0 + b1*x1
- y is the dependent variable 
- x is the independent variable (feature, there can be multiple features)
- b0 is constant (the point where line cuts y axis) 
- b1, b2, b3... will be our parameters on which the relation between x and y is defined (b1 is the slope in this case). 

Fitting the data using a straight line (best fit).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts


import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("Salary_Data.csv")
print(df.shape)
print(df.describe())
print("Number of Nan \n", df.isnull().sum())
print(df.head())
```

### Visualising the data

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

### How to find line of Best Fit?

#### - Ordinary Least Squares (OLS)
    - Draw vertical lines from each data point to out line of fit. This distance is called as *Residual*
    - Calculate this squared distance (y - y^)^2
    - Calculate the sum of all such squared distances
    - Find the line which has minimum sum, that is our line of Best Fit! 

### Training Simple Linear Regression


Note: Library for Simple Linear Regression takes care of Feature Scaling, so no need of performing Feature Scaling here

```python
X = df[['YearsExperience']]       #NOTE: input independent variable should be a 2D array, not 1D array
y = df['Salary']
X_train, X_test, y_train, y_test = tts(X,y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#predicting the Test set results
y_pred = regressor.predict(X_test)
y_pred
```

```python
#Visualising the Training set results
plt.scatter(X_train,y_train, color='red')        #scatter plot of actuall data in red color
plt.plot(X_train, regressor.predict(X_train), color='blue')    #line of fit, showing the predicted values for X_train(training set itself)
plt.title('Salary vs Experience (Training set)')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()         #this indicates the end of this plot
```

```python
#Visualising the Test set results
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')       #line of best fit, which is fit on training data
plt.scatter(X_test, y_pred, color='green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()         
```

The above visualisation shows that this linear regression line is very well fit on the data as all the predicted values are close enough to the real ground truth data points in the test set.
